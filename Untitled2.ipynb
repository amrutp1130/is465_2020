```python
import geopandas as gpd 
shapefile = './County_Boundaries_of_NJ/County_Boundaries_of_NJ.shp'
map_file = gpd.read_file(shapefile)

map_file.head()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
datafile = './covid-19-NJ.csv'
r_datafile = pd.read_csv(datafile, thousands=',')
r_datafile.head()
```

```python
map_file.plot()
```

```python
merged = map_file.merge(r_datafile, how = 'outer', left_index = True , right_index = True)
merged.head()
```

```python
variable = 'Cases'
vmin, vmax = 100,230
fig, ax = plt.subplots(1, figsize=(10, 6))
merged.plot(column=variable, cmap= 'Reds', linewidth=0.8, ax=ax, edgecolor='0.8')
ax.set_title('Covid-19 Cases in New Jersey', \
              fontdict={'fontsize': '25',
                        'fontweight' : '3'})

sm = plt.cm.ScalarMappable(cmap='Reds', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
ax.axis('off')
```
