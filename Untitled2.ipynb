```python
import geopandas as gpd 
shapefile = './County_Boundaries_of_NJ/County_Boundaries_of_NJ.shp'
map_file = gpd.read_file(shapefile)

map_file.head()
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
datafile = './covid-19-NJ.csv'
r_datafile = pd.read_csv(datafile)
r_datafile.head()
```

```python
map_file.plot()
```

```python
merged = map_file.merge(r_datafile, left_on = 'COUNTY' , right_on = 'COUNTY')
gpd.GeoDataFrame(merged, crs=map_file)
merged.head()
```
